```python
import matplotlib.pyplot as pt
```

```python
x = [2,4]
y = [4,9]
pt.plot(x,y)
pt.show()
```

```python
x = [2,6]
y = [4,9]
x1 = [8,3]
y1 = [4,7]
pt.xlabel ("time")
pt.xlabel ("speed")
pt.plot(x,y)
pt.plot(x1,y1)
pt.show()
```

```python
x = [2,6]
y = [4,9]
x1 = [8,3]
y1 = [4,7]
pt.xlabel("time")
pt.ylabel("speed")
pt.plot(x,y,label = "silver")
pt.plot(x1,y1,label = "gold")
pt.legend()
pt.show()
```

```python
x = [2,6]
y = [4,9]
x1 = [8,3]
y1 = [4,7]
pt.xlabel("time")
pt.ylabel("speed")
pt.plot(x,y,label = "silver")
pt.plot(x1,y1,label = "gold")
pt.legend()
pt.show()

```

```python
x = [2,6]
y = [4,9]
pt.xlabel("time")
pt.ylabel("speed")
pt.scatter(x,y, )


pt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
pt.show()
```

```python
x = [2,6]
y = [4,9]
x1 = [8,3]
y1 = [4,7]
pt.grid(color = 'y')
pt.xlabel("time")
pt.ylabel("speed")
pt.bar(x,y)
pt.bar(x1,y1)

pt.show()

```
